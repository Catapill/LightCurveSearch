```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from operator import itemgetter
from csv import writer
```

```python
#Loads data
df = pd.read_csv('local_reduced.csv', sep=',', header=None)
data = df.drop([0,1,2], axis=1)
label = df[0]
```

```python
#Generates noise one at a time because past adum is dumb
noise_one = np.random.normal(0,0.2,200)
noise_two = np.random.normal(0,0.2,200)
noise_three = np.random.normal(0,0.2,200)
noise_four = np.random.normal(0,0.2,200)
noise_five = np.random.normal(0,0.2,200)
noise_six = np.random.normal(0,0.2,200)
noise_seven = np.random.normal(0,0.2,200)
plt.figure(figsize=(15,10))
plt.plot(noise_one)
plt.plot(noise_two)
plt.plot(noise_three)
plt.plot(noise_four)
plt.plot(noise_five)
plt.plot(noise_six)
plt.plot(noise_seven)
plt.show()
```

```python
#Applies noise to data and adds new data to csv
with open('synth_data2.csv', 'a+', newline='') as write_obj:
    # Create a writer object from csv module
    csv_writer = writer(write_obj)
    for i, row in data.iterrows():
        data_label = label[i]

        original = row
        synth_1 = row + noise_one
        synth_2 = row + noise_two
        synth_3 = row + noise_three
        synth_4 = row + noise_four
        synth_5 = row + noise_five
        synth_6 = row + noise_six
        synth_7 = row + noise_seven

        data_label = pd.Series(data_label)
        output_1 = data_label.append(synth_1)
        output_2 = data_label.append(synth_2)
        output_3 = data_label.append(synth_3)
        output_4 = data_label.append(synth_4)
        output_5 = data_label.append(synth_5)
        output_6 = data_label.append(synth_6)
        output_7 = data_label.append(synth_7)
        
        # Add contents of list as last row in the csv file
        csv_writer.writerow(output_1)
        csv_writer.writerow(output_2)
        csv_writer.writerow(output_3)
        csv_writer.writerow(output_4)
        csv_writer.writerow(output_5)
        csv_writer.writerow(output_6)
        csv_writer.writerow(output_7)
```

```python
#plots example of new noise altered data
plt.figure(figsize=(15,10))
plt.plot(original)
plt.show()

plt.figure(figsize=(15,10))
plt.plot(synth_1)
plt.plot(synth_2)
plt.plot(synth_3)
plt.plot(synth_4)
plt.plot(synth_5)
plt.plot(synth_6)
plt.plot(synth_7)
plt.show()
```
